# LOTKA-VOLTERRA

 ## ¿Que es el modelo Lotka-Volterra?

Modelo que representa un  modelo escalatorio que se relaciona con la presentacion Presa-Depredador

Este código simula cómo interactúan dos especies a lo largo del tiempo en un modelo de tipo depredador-presa, generando un gráfico que muestra la evolución de ambas poblaciones en función del tiempo. Los parámetros a, b, c, y e determinan las tasas de crecimiento, mortalidad, y las interacciones entre las especies.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys
```

```python
# model parameters
a = 0.7; b = 0.5; c = 0.3;  e = 0.2
dt = 0.001; max_time = 100
```

```python
# initial time and populations
t = 0; x = 1.0; y = 0.5
```

```python
# empty lists in which to store time and populations
t_list = []; x_list = []; y_list = []
```

```python
# initialize lists
t_list.append(t); x_list.append(x); y_list.append(y)
```

```python
while t < max_time:
    # calc new values for t, x, y
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

 # store new values in lists
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)
```

```python
# Plot the results
p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)
```
